```python



# Function to generate assignment questions using Few-Shot Prompting
def generate_assignment_questions(input_text , logs=True):
    prompt = (
    "You are an AI assistant that generates high-quality assignment question (one ) with a focus on structured problem-solving and coding implementation.\n\n"
    "#### Question: Implement a Vacuum Cleaner Agent\n"
    "A vacuum cleaner agent operates in a 2x1 grid with two locations that may or may not contain dirt. The agent can move left, move right, and suck dirt.\n\n"
    "**Tasks:**\n"
    "a) Define a state representation for the vacuum cleaner environment.\n"
    "b) Implement a Python class to simulate the vacuum agent with appropriate methods.\n"
    "c) Define an initial state and implement a function to transition between states based on actions.\n"
    "d) Implement a goal test function that checks if all locations are clean.\n"
    "e) Extend your solution to support a larger 2x2 grid environment.\n\n"
    "Now, generate a new set of 1 assignment-style questions based on the following topic:\n\n"
    f"{input_text}\n\n"
    "**Ensure:**\n"
    "- Each task builds upon the previous one, increasing in difficulty.\n"
    "- Tasks require coding-based implementation with structured sub-parts.\n"
    "- Question and cover conceptual understanding, critical thinking, and problem-solving.\n"
    "- Higher-order question and tasks should involve optimization, performance analysis, or real-world applications."
    "- only generate 1 question keep the strict structure and dont provide Solution, No Solution, only Questions and Tasks no subtasks ,Keep output formated and structred"

    )


    # Sending request to Together API
    response = client.chat.completions.create(
        model="meta-llama/Llama-3.2-3B-Instruct-Turbo",
        # model="meta-llama/Meta-Llama-3.1-8B-Instruct-Turbo-128K",
        # model="deepseek-ai/DeepSeek-V3-Distill-Llama-70B-free",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=300,
        temperature=0.7,
        top_p=0.7,
        top_k=50,
        repetition_penalty=1,
        stop=["<|eot_id|>", "<|eom_id|>"]
    )

    ## Returning the response
    ## Streaming the response
    if logs:
      print("OLD QUESTION",input_text)
      print("NEW QUESTIONS")
      print(response.choices[0].message.content)
      print("---------------------------------------")
    return response.choices[0].message.content





# Example usage
questions_json = {
    "questions": [
        "Implement a technique to handle out-of-vocabulary (OOV) words in the transformer-based model, such as using a subword tokenizer or a learned embedding. Explain the advantages and disadvantages of each approach.",
        "Implement the ignore delete lists heuristic for planning problems. Describe how this heuristic reduces the number of states in the planning problem and how it can be used to find a solution more efficiently.",
        "Implement a basic heuristic search algorithm in Python, and demonstrate its use in a simple game-playing scenario.",
        "Implement a regression search algorithm for the blocks world problem to find a sequence of actions that leads from the initial state to the goal state"
    ]
}

improved_questions = generate_assignment_questions(questions_json["questions"][0])

```

# PDF GENERATE

```python
from reportlab.lib.pagesizes import A4
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, ListFlowable, ListItem
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_LEFT, TA_CENTER
from reportlab.lib import colors


def markdown_to_pdf(md_text, output_filename="assignment.pdf"):
    """
    Converts structured markdown text into a professional academic PDF.
    """

    # Initialize PDF document
    doc = SimpleDocTemplate(output_filename, pagesize=A4, leftMargin=50, rightMargin=50, topMargin=50, bottomMargin=50)
    elements = []  # List to store elements for the PDF

    # Define styles for different text elements
    styles = getSampleStyleSheet()
    styles.add(ParagraphStyle(name="Titlex", fontSize=16, spaceAfter=12, alignment=TA_CENTER, fontName="Helvetica-Bold"))
    styles.add(ParagraphStyle(name="Subtitlex", fontSize=13, spaceAfter=8, alignment=TA_LEFT, fontName="Helvetica-Bold"))
    styles.add(ParagraphStyle(name="Bodyx", fontSize=11, spaceAfter=6, alignment=TA_LEFT, fontName="Helvetica"))
    styles.add(ParagraphStyle(name="Bulletx", fontSize=11, leftIndent=15, spaceAfter=4, alignment=TA_LEFT))

    # Split text into lines
    lines = md_text.strip().split("\n")

    # Process Markdown Formatting
    bullet_points = []
    for line in lines:
        line = line.strip()

        if not line:
            elements.append(Spacer(1, 8))  # Add spacing between paragraphs
            continue

        if line.startswith("### "):  # Title
            elements.append(Paragraph(line[4:], styles["Titlex"]))

        elif line.startswith("#### "):  # Subtitle
            elements.append(Paragraph(line[5:], styles["Subtitlex"]))

        elif line.startswith("**") and line.endswith("**"):  # Bold Text
            elements.append(Paragraph(f"<b>{line.strip('*')}</b>", styles["Bodyx"]))

        elif line.startswith("- "):  # Bullet Point
            bullet_points.append(ListItem(Paragraph(line[2:], styles["Bulletx"])))

        elif line.startswith("a)") or line.startswith("b)") or line.startswith("c)") or \
             line.startswith("d)") or line.startswith("e)") or line.startswith("f)"):  # Numbered List
            elements.append(Paragraph(f"<b>{line}</b>", styles["Bodyx"]))

        else:  # Normal Paragraph
            elements.append(Paragraph(line, styles["Bodyx"]))

    # Add bullet points list to the document
    if bullet_points:
        elements.append(ListFlowable(bullet_points, bulletType="bullet"))

    # Build the PDF document
    doc.build(elements)
    print(f"✅ PDF successfully created: {output_filename}")


# Example Usage
generated_questions = []
for question in questions_json["questions"]:
    improved_question = generate_assignment_questions(question)
    generated_questions.append(improved_question)

# Convert list to a formatted string
formatted_text = "\n\n\n\n\n\n".join(generated_questions)

# Generate PDF with optimized formatting
markdown_to_pdf(formatted_text, "formatted_assignment.pdf")
```

```python
import json
import os
import time

# Paths for original and updated files
input_file = "/content/cleaned_output.json"
temp_file = "/content/temp_questions.json"
output_file = "/content/updated_questionsc.json"

# Check if a temporary file exists (to resume progress)
if os.path.exists(temp_file):
    with open(temp_file, "r", encoding="utf-8") as file:
        updated_questions = json.load(file)
    processed_chunks = len(updated_questions)  # Get progress
else:
    updated_questions = []
    processed_chunks = 0

# Read original questions
with open(input_file, "r", encoding="utf-8") as file:
    questions_json = json.load(file)

total_questions = len(questions_json)



# Start timing
start_time = time.time()

# Process remaining questions
for i in range(processed_chunks, total_questions):
    question = questions_json[i]

    if question["Type"] == "code":
        modified_question = generate_assignment_questions(question.copy())
        updated_questions.append(modified_question)
    else:
        updated_questions.append(question)  # Keep non-code questions unchanged

    # Save progress every 10 iterations
    if i % 10 == 0 or i == total_questions - 1:
        with open(temp_file, "w", encoding="utf-8") as temp:
            json.dump(updated_questions, temp, indent=4)

        # Progress and time estimation
        elapsed_time = time.time() - start_time
        completed_percent = (i + 1) / total_questions * 100
        avg_time_per_item = elapsed_time / (i + 1 - processed_chunks)
        remaining_time = (total_questions - (i + 1)) * avg_time_per_item

        print(f"\rProgress: {completed_percent:.2f}% | Time Elapsed: {elapsed_time:.2f}s | Estimated Time Left: {remaining_time:.2f}s", end="")

# Final save to output file
with open(output_file, "w", encoding="utf-8") as file:
    json.dump(updated_questions, file, indent=4)

# Cleanup temporary file
os.remove(temp_file)

# Final timing
total_time = time.time() - start_time
print(f"\n✅ Updated questions saved to {output_file} in {total_time:.2f} seconds.")
```

```python
import json
import os
import multiprocessing
import re
from tqdm import tqdm  # Import tqdm for progress bar

# Paths for original and updated files
input_file = "/content/drive/MyDrive/ThesisDataRaw/AI-Book/chunks/metaXmerged_flattened_questions.json"
output_file = "/content/drive/MyDrive/ThesisDataRaw/AI-Book/chunks/Enhanced-metaXmerged_flattened_questions.json"

pattern = r"\*\*.*?\*\*|\n\n"

# Read original questions
with open(input_file, "r", encoding="utf-8") as file:
    questions_json = json.load(file)

# User-defined stop point
# stop_at = 10  # Change this as needed
stop_at = len(questions_json)  # Change this as needed

# Function to modify a question
def process_question(q):
    if q.get("Type", "").lower() == "code":
        modified_question = generate_assignment_questions(q["Question"], False)
        q_updated = q.copy()
        q_updated["Question"] = re.sub(pattern, '', modified_question).strip()
        return q_updated
    return q  # Skip non-code questions

# Split work into chunks for parallel processing
num_workers = min(multiprocessing.cpu_count(), 4)  # Use up to 4 cores
with multiprocessing.Pool(num_workers) as pool:
    updated_questions = list(
        filter(None, tqdm(pool.imap(process_question, questions_json[:stop_at]),
                          total=stop_at, desc="Processing Questions"))
    )

# Save processed data to a new JSON file
with open(output_file, "w", encoding="utf-8") as file:
    json.dump(updated_questions, file, indent=4)

# Display summary
print("\nSummary:")
print(f"Total Questions: {len(questions_json)}")
print(f"Code Questions Processed: {len(updated_questions)}")
print(f"Processed data saved to {output_file}")
```
